```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('netflix_titles.csv')

# Show the first 5 rows
df.head()
```

```python
sns.countplot(data=df, x='type', palette='pastel')
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='barh', color='skyblue')
plt.title("Top 10 Countries")
plt.show()
```

```python
from collections import Counter
genres = ','.join(df['listed_in'].dropna()).split(',')
genres_count = Counter([genre.strip() for genre in genres])
top_genres = dict(genres_count.most_common(10))

plt.barh(list(top_genres.keys()), list(top_genres.values()), color='orange')
plt.title("Top 10 Genres on Netflix")
plt.show()
```
